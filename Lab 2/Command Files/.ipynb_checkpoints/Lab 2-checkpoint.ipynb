```python
import pandas as pd
import numpy as np
import re
import datetime

bill = pd.read_csv("billboard.csv", encoding= 'unicode_escape')
bill.head()

```

```python
#bill.columns
bill=(bill.melt(id_vars=["year","artist.inverted","track","genre","date.entered","date.peaked","time"],var_name=["week"],value_name="rank"))
bill.head()
```

```python
bill.update(pd.DataFrame({"week": np.ravel([list(map(int, re.findall(r'\d',i))) for i in bill["week"]])}))
bill.head()
```

```python
#bill.update(pd.DataFrame({"week": map(int, i) for i in bill["week"]}))
#bill.head()
```

```python
bill['date.entered'] =  pd.to_datetime(bill['date.entered'], format='%Y/%m/%d')
bill = (bill.assign(date=[bill['date.entered'][i]+datetime.timedelta(weeks=bill["week"][i][0]-1) for i in range(len(bill["week"]))])
       .drop(['date.entered','date.peaked','week'], axis=1)
       .sort_values('artist.inverted', ascending=True)
       .reset_index(drop=True))
bill.head()
```

```python
labels, unique = pd.factorize(list(bill["track"]))
bill["id"]= labels + 1
bill.reset_index(drop=True)
bill.head()
```

```python
rank = bill.filter(["id", "date", "rank"]).dropna()
rank = rank.sort_values(by=['id','date']).reset_index(drop=True)
song= bill.filter(["id", "artist.inverted", "track","time"])
song = song.drop_duplicates('id').reset_index(drop=True)

```
