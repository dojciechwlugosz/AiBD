```python
import pandas as pd
import numpy as np
import re
import datetime
import numpy as np
import matplotlib.pyplot as plt

rank = pd.read_csv("../Analysis Data Files/rank.csv", encoding= 'unicode_escape', error_bad_lines=False)
song = pd.read_csv("../Analysis Data Files/song.csv", encoding= 'unicode_escape', error_bad_lines=False)
billboard = pd.read_csv("../Original Data/billboard.csv", encoding= 'unicode_escape', error_bad_lines=False)
```

```python
plot_data = pd.DataFrame({
    "billboard.csv":[billboard.shape[1]],
    "song.csv":[song.shape[1]],
    "rank.csv":[rank.shape[1]]},
    index = [" "]
)

plot_data.plot(kind="bar", figsize=(10,15))
plt.title("Number of columns in datasets")
```

```python
plot_data2 = pd.DataFrame({
    "billboard.csv":[billboard.shape[0]],
    "song.csv":[song.shape[0]],
    "rank.csv":[rank.shape[0]]},
    index = [" "]
)

plot_data2.plot(kind="bar", figsize=(10,15))
plt.title("Number of raws in datasets")
```
