```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

path = "~/laboratorium-6-dojciechwlugosz/Analysis Data/15_ZACHODNIOPOMORSKIE.csv"
df = pd.read_csv(path) 
df_without_NaN = pd.read_csv(path) 
df_without_NaN.dropna(subset = ['Wiek kupującego'], inplace=True)
df_without_bd = df[df["Płeć kupującego"] != 'bd.']
```

```python
hist1 = pd.DataFrame({'Dni od zakupu':sorted(df['Dni od zakupu'].unique().tolist()), 'Liczba probek':df.groupby(['Dni od zakupu']).size().tolist()})
print(hist1)
hist1.plot.bar(x='Dni od zakupu', y='Liczba probek', figsize=(15,10), color = ['lightpink'])
plt.grid(True)
plt.xticks(rotation=0)
plt.xlabel('Dni od zakupu')
plt.ylabel('Liczba probek')
plt.title('Histogram ilości dni od zakupu')
plt.savefig("/home/wojtek/laboratorium-6-dojciechwlugosz/Documents/Dni_od_zakupu.png" , transparent=True)
```

```python
# Histrogram wieku kupujacego - z usunietymi wartosciami NaN

hist2 = pd.DataFrame({'Wiek kupującego':sorted(df_without_NaN['Wiek kupującego'].unique().tolist()), 'Liczba probek':df_without_NaN.groupby(['Wiek kupującego']).size().tolist()})

hist2.plot.bar(x='Wiek kupującego', y='Liczba probek', figsize=(25,20), color = ['gold'])
plt.grid(True)
plt.xticks(rotation=0)
plt.xlabel('Wiek kupującego')
plt.ylabel('Liczba probek')
plt.title('Histogram wieku kupującego - gdy wystepuja dane')
plt.savefig("~/laboratorium-6-dojciechwlugosz/Documents/Wiek_kupujacego_bez_NaN.png" , transparent=True)
```

```python
hist3 = pd.DataFrame({'Marka':sorted(df['Marka'].unique().tolist()), 'Liczba probek':df.groupby(['Marka']).size().tolist()})
print(hist3)
hist3.plot.bar(x='Marka', y='Liczba probek', figsize=(15,10), color = ['aqua'])
plt.grid(True)
plt.xticks(rotation=0)
plt.xlabel('Marka')
plt.ylabel('Liczba probek')
plt.title('Histogram marki')
plt.savefig("~/laboratorium-6-dojciechwlugosz/Documents/Marka.png" , transparent=True)
```

```python
# Histogram plci kupujacego - gdy wystepuja dane

hist4 = pd.DataFrame({'Płeć kupującego':sorted(df_without_bd['Płeć kupującego'].unique().tolist()), 'Liczba probek':df_without_bd.groupby(['Płeć kupującego']).size().tolist()})
print(hist4)
hist4.plot.bar(x='Płeć kupującego', y='Liczba probek', figsize=(15,10), color = ['turquoise'])
plt.grid(True)
plt.xticks(rotation=0)
plt.xlabel('Płeć kupującego - gdy wystepuja dane')
plt.ylabel('Liczba probek')
plt.title('Histogram plci - gdy wystepuja dane')
plt.savefig("~/laboratorium-6-dojciechwlugosz/Documents/Plec_kupujacego_bez_bd.png" , transparent=True)
```
